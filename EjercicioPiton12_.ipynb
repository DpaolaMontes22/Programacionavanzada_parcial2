```python
appat="" 
apmat="" 
nombr="" 
año=""  
mes=""   
dia="" 
sexo=""  
est=""  
part1="" 
part2="" 
part3="" 
part4="" 
curp=""  


def nomComp(apt, amt, nom):
    voc=["A","E","I","O","U"]
    bueno=True
    i=1
    resultado=""
    while bueno and i<len(apt):
        letra=apt[i]
        v=0
        encuentra=True
        while v<len(voc) and encuentra:
            if letra==voc[v]:
                resultado=letra
                encuentra=False
                bueno=False
            v=v+1
        i=i+1
    sum1=apt[0]+resultado+amt[0]+nom[0]
    return sum1.upper()

def fecha(a, m, d):
    mescor="0"
    diacor="0" 
    
    if len(m)==1: mescor=mescor+m 
    
    elif len(m)==2: mescor=m
    
  
    if len(d)==1: diacor=diacor+d 
    
    elif len(d)==2: diacor=d 
    
    part_2=a[2]+a[3]+mescor+diacor
    return part_2


def EF(stdo):
    es=["AS","BC","BS","CC","CL","CM","CS","CH","DF","DG","GT","GR","HG","JC","MC","MN","MS","NT","NL","OC","PL","QT","QR","SP","SL","SR","TC","TS","TL","VZ","YN","ZS"]
    num=int(stdo)
    num=num-1
    return es[num]

def conson(pat, mat, nom):
    abc=["B","C","D","F","G","H","J","K","L","M","N","Ñ","P","Q","R","S","T","V","W","X","Y","Z"]
    L1=""
    L2=""
    L3=""
   
    bueno1=True
    a=1
    while bueno1 and a<len(pat):
        letra=pat[a]
        i=0
        encuentra=True
        while i<len(abc) and encuentra:
            if letra==abc[i]:
                L1=letra
                encuentra=False
                bueno1=False
            i=i+1
        a=a+1
  
    bueno2=True
    b=1
    while bueno2 and b<len(mat):
        letra=mat[b]
        i=0
        encuentra=True
        while i<len(abc) and encuentra:
            if letra==abc[i]:
                L2=letra
                encuentra=False
                bueno2=False
            i=i+1
        b=b+1
 
    bueno3=True
    c=1
    while bueno3 and c<len(nom):
        letra=nom[c]
        i=0
        encuentra=True
        while i<len(abc) and encuentra:
            if letra==abc[i]:
                L3=letra
                encuentra=False
                bueno3=False
            i=i+1
        c=c+1
    return L1+L2+L3
    


print("PULSE ENTER AL TERMINAR DE INGRESAR LOS DATOS PARA CONTINUAR")
print("INGRESE SU APELLIDO PATERNO")
appat=input()
print("INGRESE SU APELLIDO MATERNO")
apmat=input()
print("INGRESE SU NOMBRE O NOMBRES")
nombr=input()
print("INGRESE AÑO DE NACIMIENTO")
año=input()
print("INGRESE NÚMERO DEL MES DE NACIMIENTO")
mes=input()
print("INGRESE NÚMERO DEL DÍA DE NACIMIENTO")
dia=input()
print("INGRESE SU SEXO")
print("H si es Hombre o M si es Mujer ")
sexo=input()

print("ESCRIBE EL NÚMERO DE TU ENTIDAD FEDERATIVA")
print("1- Aguascalients        2- Baja California Norte")
print("3- Baja California sur  4- Campeche")
print("5- Coahuila             6- colima")
print("7- Chiapas              8- Chihuahua")
print("9- CDMX                 10-Durango")
print("11-Guanajuato           12-Guerrero")
print("13-Hidalgo              14-Jalisco")
print("15-EDOMEX               16-Michoacán")
print("17-Morelos              18-Nayarit")
print("19-Nuevo León           20-Oaxaca")
print("21-Puebla               22-Queretaro")
print("23-Quintana Roo         24-San Luis Potosí")
print("25-Sinaloa              26-Sonora")
print("27-Tabasco              28-Tamaulipas")
print("29-Tlaxcala             30-Veracruz")
print("31-Yucatán              32-Zacatecas")
est=input()

part1=nomComp(appat.upper(),apmat.upper(),nombr.upper())
part2=fecha(año,mes,dia)
part3=EF(est)
part4=conson(appat.upper(),apmat.upper(),nombr.upper())

from random import sample
numeros=sample([x for x in range(0,9)],2)
n1=str(numeros[0])
n2=str(numeros[1])

curp=part1+part2+sexo.upper()+part3+part4+n1+n2

print(curp)
```
