```python
def torresHanoi (n, inicio,mitad,fin):
    if n==1:
        print(f"el disco se mueve de {inicio} a {mitad}")
        return
    torresHanoi(n-1,inicio,fin,mitad)
    print(f"el disco se mueve {n} de {inicio} a {mitad}")
    torresHanoi(n-1,fin, mitad, inicio)
    
ciclo=True
while ciclo:
 try:
     pass 
     print("¿Cuantos desea mover")
     numero = int(input())
     if numero>2 and numero<16:
         ciclo=False
          
 except Exception:
     pass
     print("No es positivo!")  

print()
torresHanoi(numero,"A","B","C")
```
